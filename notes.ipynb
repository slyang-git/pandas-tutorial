```python
data = {
    "Name": ["Alice", "Bob", "Alice", "Bob", "Charlie"],
    "Month": ["Jan", "Jan", "Feb", "Feb", "Jan"],
    "Sales": [100, 200, 150, 300, 250],
}
df2 = pd.DataFrame(data)
df2
```

```python
pivot_table = df2.pivot_table(
    values="Sales", index="Name", columns="Month", aggfunc="sum"
)
pivot_table
```
